```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
import scipy

df = pd.read_csv("/home/panu/Desktop/aibe/data/raw/marketing_campaign.csv" , sep = ";")  # Replace with cleaned path
df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
median_income = df['Income'].median()
df['Income'] = df['Income'].fillna(median_income)
df.isnull().sum().sort_values(ascending=False)
#Filling the missing values using - Median Imputation
```

```python
df['Age'] = 2025 - df['Year_Birth']
df['Engagement_Score'] = 100 - df['Recency']  # Higher means more recently active

df.describe()
```

```python
campaign_cols = ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']
df['Total_Accepted_Campaigns'] = df[campaign_cols].sum(axis=1)

df['Campaign_Response_Rate'] = df['Total_Accepted_Campaigns'] / len(campaign_cols)

df['Responded_Last'] = df['Response']

df.head()

#Campaign performance features
```

```python
spend_cols = ['MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts','MntGoldProds']
df['Total_Spending'] = df[spend_cols].sum(axis=1)

df['Luxury_Spending_percentage'] = (df['MntGoldProds'] / df['Total_Spending'])* 100


df.head()

#Product Affinity Features
```

```python
purchase_cols = ['NumDealsPurchases','NumWebPurchases','NumCatalogPurchases','NumStorePurchases']
df['Total_Purchases'] = df[purchase_cols].sum(axis=1)

df['Web_Purchase_Ratio'] = df['NumWebPurchases'] / df['Total_Purchases']

df['Catalog_Loyalty'] = df['NumCatalogPurchases'] / df['Total_Purchases']

df['Online_Engagement'] = df['NumWebPurchases'] + df['NumWebVisitsMonth']

df.head()

#Channel behaviour & preference
```

```python
df['Children'] = df['Kidhome'] + df['Teenhome']
df['Family_Size'] = df['Children'] + 2
df.head()

#Family Demographics (Assuming 2 Adults)


```

```python
income_threshold = df['Income'].median()
df['High_Income'] = (df['Income'] > income_threshold).astype(int)
spending_threshold = df['Total_Spending'].median()
df['High_Spender'] = (df['Total_Spending'] > spending_threshold).astype(int)
df['ROI_Flag'] = ((df['High_Income'] == 1) & (df['High_Spender'] == 0)).astype(int)
df.head()
#High-Income buyer flags
```

```python
from datetime import datetime

# Drop ID and constant-value columns
df.drop(columns=["ID", "Z_CostContact", "Z_Revenue"], inplace=True)

# Convert 'Dt_Customer' to datetime and calculate Customer Tenure
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
df['Customer_Tenure'] = (datetime(2025, 7, 14) - df['Dt_Customer']).dt.days
df.drop(columns=['Dt_Customer'], inplace=True)

# Create aggregated features
spend_cols = ['MntWines', 'MntFruits', 'MntMeatProducts',
              'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
df['TotalSpend'] = df[spend_cols].sum(axis=1)

df['TotalChildren'] = df['Kidhome'] + df['Teenhome']
df['TotalAcceptedCmp'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 
                             'AcceptedCmp4', 'AcceptedCmp5']].sum(axis=1)

# One-hot encode categorical columns
df = pd.get_dummies(df, columns=['Education', 'Marital_Status'], drop_first=True)

# Drop redundant columns
df.drop(columns=spend_cols + ['Kidhome', 'Teenhome',
                              'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3',
                              'AcceptedCmp4', 'AcceptedCmp5'], inplace=True)

# Final check
print("✅ Preprocessing complete!")
print("🔎 Final shape:", df.shape)
df.head()
```
